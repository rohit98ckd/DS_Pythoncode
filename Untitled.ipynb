```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AAPL.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
data = df.set_index(['Date'])
```

```python
data
```

```python
data['Close'].plot(figsize=(18,6))
```

```python
data.resample(rule= 'A').min()  # minimum in the each year
```

```python
data.resample(rule= 'A').min().plot()
```

```python
data.resample(rule= 'A').max().plot()
```

```python
data.resample(rule= 'A').min()['Close'].plot()
```

```python
data.resample(rule= 'A').max()['Close'].plot()
```

```python
data.resample(rule= 'BA').max()
```

```python
data.resample(rule= 'BA').max()
```

```python
data.resample(rule= 'BA').max().plot(kind= "bar")
```

```python
data['Close'].resample(rule= 'A').max().plot(kind="bar",figsize=(18,6))
```

```python
data['Close'].resample(rule= 'M').max().plot(kind="bar",figsize=(18,6))
```

```python
data['Close'].rolling(30).mean().plot(figsize=(18,6))
```

```python
plt.figure(figsize=(12,4))
data.Close.plot(label="org")
plt.title("original series")

for i in range(0,365,10):                       # to find optimal windaow size
    plt.figure(figsize=(12,4))
     
    data['Close'].rolling(i).mean().plot(label=str(i))     # roling is nothing but moving average
    plt.title("MA "+str(i))
    plt.legend(loc='best')
    plt.show()
```

```python
#EWMA -- Exponential weighted moving average
```

```python
data['Close'].ewm(alpha=0.1, adjust=False).mean()
```

```python
data['Close'].ewm(alpha=0.1, adjust=False).mean().plot(figsize=(18,6))
```

```python
data['Close'].ewm(span= 30).mean().plot(figsize=(18,6))
```
